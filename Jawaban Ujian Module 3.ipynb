## 9. Use K Fold (K=5,shuffle=True,random_state=42), every loop use RandomOverSampler for the training data, and then use GradientBoostingClassifier for the model

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
data = dfpca.drop(['Class'], axis=1)
target = dfpca['Class']
```

```python
def calc_train_error(X_train, y_train, model):
    # Return in-sample error for already fit model
    predictions = model.predict(X_train)
    predictProba = model.predict_proba(X_train) # buat roc_auc disini
    matt = matthews_corrcoef(y_train, predictions)
    f1 = f1_score(y_train, predictions, average='macro')
    report = classification_report(y_train,predictions)
    roc_auc = roc_auc_score(y_train, predictProba[:,1])
    accuracy = accuracy_score(y_train, predictions)
    confMatrix = confusion_matrix(y_train, predictions)
    logloss = log_loss(y_train, predictProba)
    return {
        'report' : report,
        'matthew' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'confusion' : confMatrix,
        'logloss' : logloss
    }

def calc_validation_error(X_test, y_test, model):
    # Return out-of-sample error for already fit model
    predictions = model.predict(X_test)
    predictProba = model.predict_proba(X_test)
    matt = matthews_corrcoef(y_test, predictions)
    f1 = f1_score(y_test, predictions, average='macro')
    report = classification_report(y_test,predictions)
    roc_auc = roc_auc_score(y_test, predictProba[:,1])
    accuracy = accuracy_score(y_test, predictions)
    confMatrix = confusion_matrix(y_test, predictions)
    logloss = log_loss(y_test, predictProba)
    return {
        'report' : report,
        'matthew' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'confusion' : confMatrix,
        'logloss' : logloss
    }

def calc_metrics(X_train, y_train, X_test, y_test, model):
    # Fits model and returns the classification matrix for in-sample error and out-of-sample error
    model.fit(X_train, y_train)
    train_error = calc_train_error(X_train, y_train, model)
    validation_error = calc_validation_error(X_test, y_test, model)
    return train_error, validation_error
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score, matthews_corrcoef, f1_score, log_loss
from imblearn.over_sampling import RandomOverSampler
```

```python
train_errors = []
validation_errors = []

for train_index, val_index in kf.split(data,target):
    #split data
    X_train, X_val = data.iloc[train_index], data.iloc[val_index]
    y_train, y_val = target.iloc[train_index], target.iloc[val_index]
    
    # RANDOM OVER SAMPLER
    ros = RandomOverSampler()
    X_ros, y_ros = ros.fit_sample(X_train, y_train)
    
    #instantiate model
    gbc = GradientBoostingClassifier()
    
    #calculate errors
    train_error, val_error = calc_metrics(X_ros, y_ros, X_val, y_val, gbc)
    
    #append to appropriate list
    train_errors.append(train_error)
    validation_errors.append(val_error)
```

## 10. Print the Model Evaluation like below from the K Fold 

```python
listItem = []
listIndex = []

for i,t,v in zip(range(1,len(train_errors)+1),train_errors, validation_errors):
    listItem.append([t['accuracy'], v['accuracy'], t['roc'], v['roc'], t['f1'], v['f1'], t['matthew'], v['matthew'], t['logloss'], v['logloss']])
    listIndex.append(i)
                
dfEvaluate = pd.DataFrame(listItem ,columns=['Train Accuracy', 'Test Accuracy', 'Train ROC AUC', 'Test ROC AUC', 'Train F1 Score', 'Test F1 Score', 'Train Matthews Corr Coef', 'Test Matthews Corr Coef', 'Train Log Loss', 'Test Log Loss'],index = listIndex)
dfEvaluate
```

```python
print("Avg Accuracy Train : {}".format((list(np.mean(listItem, axis=0)))[0]))
print("Avg Accuracy Validation : {}".format((list(np.mean(listItem, axis=0)))[1]))
print("Avg ROC AUC Train : {}".format((list(np.mean(listItem, axis=0)))[2]))
print("Avg ROC AUC Validation : {}".format((list(np.mean(listItem, axis=0)))[3]))
```
